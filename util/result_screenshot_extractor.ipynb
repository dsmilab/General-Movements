```python
import taiko as tk
import pandas as pd
```

```python
SONG = [3, 3, 3, 4]
df = pd.DataFrame(columns=['drummer_id', 'song_id', 'performance_order', 'score',
                           'perfect', 'good', 'miss', 'max_combo', 'total_combo'])
id_ = 0
for drummer in range(8):
    for song in range(4):
        for order in range(SONG[song]):
            arr = tk.result_board(drummer + 1, song + 1, order + 1).img_screenshot
            df.loc[id_] = [drummer + 1, song + 1, order + 1] + arr
            id_ += 1
```

```python
df.to_csv('CSV/taiko_result.csv', index=False)
df
```
