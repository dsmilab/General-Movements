```python
import taiko as tk
from tqdm import tqdm
# only doraemon
DRUMMERS = 8
SONG = 1
P_COUNT = 3
```

```python
for drummer in tqdm(range(DRUMMERS)):
    for order_id in range(P_COUNT):
        event_primitive_df = tk.get_performance(drummer + 1, SONG, order_id + 1).event_primitive_df
        filename = '%d-%d-%d-%s%s' % (drummer + 1, SONG, order_id + 1, 'y', 'y')
        event_primitive_df.to_csv('CSV/event_primitive/' + filename + '.csv', index=False)
```
