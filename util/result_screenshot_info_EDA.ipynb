```python
import taiko as tk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
result_df = pd.read_csv('CSV/taiko_result.csv')
result_df['ok'] = result_df['perfect'] + result_df['good']
result_df['total_beat'] = result_df['perfect'] + result_df['good'] + result_df['miss']
result_df['ok_ratio'] = result_df['ok'] / result_df['total_beat']
result_df['perfect_ratio'] = result_df['perfect'] / result_df['total_beat']
result_df
```

```python
for col in ['score', 'perfect', 'ok', 'miss', 'max_combo', 'total_stream']:
    sns.catplot(x="drummer_id", y=col,
                capsize=.3, aspect=2, height=6,
                kind="point", data=result_df, hue='song_id')
    plt.show()
    plt.figure(figsize=(15, 8))
    sns.boxplot(x="drummer_id", y=col, data=result_df, hue='song_id')
    plt.show()
```

```python
for col in ['ok_ratio', 'perfect_ratio']:
    sns.catplot(x="drummer_id", y=col,
                capsize=.3, aspect=2, height=6,
                kind="point", data=result_df, hue='song_id')
    plt.show()
    plt.figure(figsize=(15, 8))
    sns.boxplot(x="drummer_id", y=col, data=result_df, hue='song_id')
    plt.show()
```
