```python
import taiko as tk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
result_df = pd.DataFrame(
    columns=['test_who', 'song_id', 'test_order', 'f1_score', 'mode'])
MODE = ['one-to-one', 'rest-to-one', 'all-to-one']
id_ = 0

for song in range(4):
    model = tk.LGBM(song + 1)

    for who in tqdm(range(8)):
        for mode in MODE:
            res = model.run(who + 1, mode=mode)
            for key, val in res.items():
                result_df.loc[id_] = [who + 1, song + 1, key, val, mode]
                id_ += 1
result_df
```

```python
result_df.to_csv('CSV/lgbm_result.csv', index=False)
```
