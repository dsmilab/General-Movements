```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
result_df = pd.read_csv('CSV/lgbm_result.csv')
result_df
```

```python
sns.catplot(x="test_who", y="f1_score",
            capsize=.3, aspect=2,
            
            kind="point", data=result_df, hue='song_id', col='mode', row='feat', vmax=1)
plt.show()
```

```python
sns.catplot(x="test_who", y="f1_score", capsize=.3, aspect=2,
            kind="point", data=result_df, hue='feat', row='song_id', col='mode', vmax=1)
plt.show()

sns.catplot(x="test_who", y="f1_score", aspect=2,
            kind="box", data=result_df, hue='feat', row='song_id', col='mode')
plt.show()
```

```python
plt.figure(figsize=(12, 20))
sns.boxplot(y='test_who', x='f1_score', data=result_df, hue='feat', orient='h')
plt.show()
```
