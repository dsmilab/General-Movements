```python
from apps.lib import Sensor
from apps.score import ScoreBoard
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
sensor_data = Sensor(verbose=0)
```

```python
scoreboard = ScoreBoard(sensor_data.drummer_df)
```

```python
def plot_scores(who_id, song_id):
    plt.figure(figsize=(15, 5))
    
    pf_total = 3
    if song_id == 4:
        pf_total = 4
        
    for pf_order in range(1, pf_total + 1):
        plt.plot(scoreboard.get_scores(who_id, song_id, pf_order))

        plt.xlabel('frame')
        plt.ylabel('score') 
        plt.title("Score, (who=%d song=%d)" % (who_id, song_id))
        
#         plt.savefig("Score, (who=%d song=%d).jpg" % (who_id, song_id))
    plt.legend(labels=['order %d' % pf_order for pf_order in range(1, pf_total + 1)], loc='best')
    plt.show()
```

# score visualization

## song 1 (夢をかなえてドラえもん)

```python
for who in range(8):
    plot_scores(who + 1, 1)
```

## song 2 (ウィーアー！)

```python
for who in range(8):
    plot_scores(who + 1, 2)
```

## song 3 (六兆年と一夜物語)

```python
for who in range(8):
    plot_scores(who + 1, 3)
```

## song 4 (残酷な天使のテーゼ)

```python
for who in range(8):
    plot_scores(who + 1, 4)
```
