```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
result_df = pd.read_csv('CSV/song1_result-yy.csv')
result_df
```

```python
ming_df = result_df[(result_df['train_who'] == result_df['test_who'])
                    & (result_df['train_who'] == result_df['valid_who'])]
```

```python
ming_df
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(y='f1_score', x='train_who', data=ming_df)
plt.show()
```

```python
df = result_df.groupby(['train_score', 'test_score'])['f1_score'].mean().reset_index()
plt.figure(figsize=(20, 10))
sns.heatmap(df.pivot('train_score', 'test_score', 'f1_score'), annot=True, fmt=".2f", cmap="YlGnBu", vmax=1.0, vmin=0.5)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.catplot(y='f1_score', x='train_who', data=result_df, hue='test_who', row='valid_who', kind='bar', height=6, aspect=2)
plt.show()
```
